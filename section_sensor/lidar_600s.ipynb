```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv("sensor_data_600.txt", delimiter = " ",
                    header = None, names = ("data", "time", "ir", "lidar"))

data["lidar"].hist(bins = max(data["lidar"]) - min(data["lidar"]), align = 'left')
plt.show()
```

```python
data.lidar.plot()
plt.show()
```

```python
data["hour"] = [e//10000 for e in data.time]
d = data.groupby("hour")
print(d)
d.lidar.mean().plot()
plt.show()
```

```python
d.lidar.get_group(6).hist()
d.lidar.get_group(14).hist()
plt.show()
```

```python
each_hour = {i:d.lidar.get_group(i).value_counts().sort_index() for i in range(24)}
freqs = pd.concat(each_hour, axis = 1)
freqs = freqs.fillna(0)
probs = freqs/len(data)
probs #同時確率
```

```python
  import seaborn as sns

  sns.heatmap(probs)
  plt.show()
```

```python
p_t = pd.DataFrame(probs.sum()) #行ベクトルの総和をとってDataFrameに
p_t.plot()
p_t.transpose()
```

```python
p_t.sum()
```

```python
p_z = pd.DataFrame(probs.transpose().sum()) 列ベクトルの総和
p_z.plot()
p_z.transpose()
```

```python
 cond_z_t = probs/ p_t[0]
 print(p_t[0])
 cond_z_t
```

```python
 (cond_z_t[6]).plot.bar(color="blue",alpha = 0.5)
 (cond_z_t[14]).plot.bar(color="orange",alpha = 0.5)
 plt.show()
```

```python
cond_t_z  = probs.transpose()/probs.transpose().sum()

print("P(z=630=",p_z[0][630])
print("P(t=13)=",p_t[0][13])
print("P(t=13|z = 630=",cond_t_z[630][13]*p_z[0][630]/p_t[0][13])
print("answerP(z=630|t=13)=",cond_z_t[13][630])
```

```python
def bayes_estimation(sensor_value, current_estimation):
    new_estimation = []
    for i in range(24):
        new_estimation.append(cond_z_t[i][sensor_value]*current_estimation[i])
    
    return new_estimation/sum(new_estimation) #正規化
```

```python
estimation = bayes_estimation(630,p_t[0]) #p(t|z=630)を計算
plt.plot(estimation)
```

```python
value_5 = [630,632,636] #5時台に取れたデータ3つを格納

estimation = p_t[0]
for v in value_5:
    estimation = bayes_estimation(v,estimation)#p(t|z=630)*p(t|z=632)*p(t|z=636)を計算

plt.plot(estimation) #これらが現れた時間を調べると5時台なので概ねあっている．
```

```python
value_11 = [617,624,619] #11時台にとれた３つのデータ

estimation = p_t[0]
for v in value_11:
    estimation = bayes_estimation(v,estimation)

plt.plot(estimation) #ちょっとずれている．
```
